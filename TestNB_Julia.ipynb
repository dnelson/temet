```julia
function is_divisble(n, i)
    return n % i == 0
end

function is_prime(n::Int64)
    if is_divisble(n, 2)
        return false
    end
 
    # initialize a counter variable
    i = 3
 
    while i <= sqrt(n)
        if is_divisble(n, i)
            return false
        end
 
        i += 2
    end
 
    return true
end
```

```julia
is_prime(97)
```

```julia
is_prime(1234567)
```
